```python
import numpy as np
import pandas as pd

edge=pd.read_excel('../data/distance.xlsx')
node=pd.read_excel('../data/序列数据.xlsx')
node
```

```python
edge
```

```python
strain_name=list(node['viruStrain'])
seq_to_idx={strain_name:i for i,strain_name in enumerate(strain_name)}
seq_to_idx
```

```python
# seq=name.join(seq)
node=node.reset_index()
con1=pd.merge(edge,node,left_on='virusStrain1',right_on='viruStrain')
con1=con1.reset_index()
con1
```

```python
con2=pd.merge(con1,node,left_on='virusStrain2',right_on='viruStrain')

con2=pd.concat([con2,pd.DataFrame(columns=['indexon'])])
con2
```

```python
con2.rename(columns={'viruStrain_x':'virustrain','index_x':'virus_num','index_y':'serum_num','seq_x':'virus_seq','seq_y':'serum_seq','year_x':'virus_year','year_y':'serum_year'},inplace=True)
con2
```

```python
distance=con2[['virusStrain1','virus_num','virusStrain2','serum_num','distance','virus_seq','serum_seq','virus_year','serum_year']]
```

```python
distance
```

```python
titer=distance[['virus_num','serum_num','distance']]
titer['distance'].replace('<', '',inplace=True, regex=True)
titer.to_excel('distance_graph.xls')
```
