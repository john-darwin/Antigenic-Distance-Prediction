```python
import pandas as pd
G=pd.read_excel('D:\Code\pycharm\AANE对比实验\补全之后的抗原距离.xlsx')
M=pd.read_excel('D:\Code\pycharm\AANE对比实验\data/titre.xlsx')
G
```

```python
M['titer'] = M['titer'].str.replace('\D+', '')
M['titer']=pd.to_numeric(M['titer'], errors='coerce').fillna('0').astype('int32')
M
```

```python
import math
for index, row in G.iterrows():
        K=M[M['virus_num']==row['virus_num']]
        K=K[K['serum_num']==row['serum_num']].max() # tij
        S=M[M['virus_num']==row['serum_num']]
        S=S[S['serum_num']==row['serum_num']].max() # tjj
        if math.isnan(S['titer']) or math.isnan(K['titer']):
            K=M[M['virus_num']==row['virus_num']]
            K=K[K['serum_num']==row['serum_num']].max() # tji
            S=M[M['virus_num']==row['virus_num']]
            S=S[S['serum_num']==row['virus_num']].max() # tjj
        if S['titer']>K['titer']:
            G['distance2'][index]=math.log2(S['titer']/K['titer'])
        else:
            G['distance2'][index]=math.log2(K['titer']/S['titer'])
```
